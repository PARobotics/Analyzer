# Move Control Data Analyzer
Parses and plots the data from the drive train in order to fine tune its PID parameters

## Usage

* Set `dataUrl` to a link which points to the raw txt file you want to process. 
* Set `plotX`, `plotY`, and `plotR` to customize the output style

Run the parser by clicking in the cell with code, and pressing the "run" button (located in the row of icons, it looks like a play/pause symbol).


```python
%matplotlib inline   
import urllib.request
import numpy as np
import matplotlib.pyplot as plt

dataUrl = "https://raw.githubusercontent.com/PARobotics/Analysis-Data/master/2017-2018/October/6/1.txt"

plotX = 1
plotY = 1
plotR = 1

lines = []

with urllib.request.urlopen(dataUrl) as url:
    s = url.read().decode("utf8").split("\n")   
    
for line in s:
    while "  " in line:
        line = line.replace("  ", " ")
    if "[MOVE] " in line and len(line.split(" ")) == 7:
        line = line.replace("[MOVE] ", "")
        lines.append(line)
        
plt.figure(figsize = (16,12))
    
x = []
y = []
r = []
    
for line in lines:
    line = line.replace("\r", "").split(" ")
    if plotX: x.append(float(line[1]))
    if plotY: y.append(float(line[0]))
    if plotR: r.append(float(line[2]))
    
if plotX: plt.plot(x, label = "Lateral Movement")
if plotY: plt.plot(y, label = "Vertical Movement")
if plotR: plt.plot(r, label = "Rotation")
    
plt.plot([0, len(lines)], [0, 0], "--")
    
plt.legend()
plt.show()
```
